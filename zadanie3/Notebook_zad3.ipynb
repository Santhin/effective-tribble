```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
ls
```

```python
df_learn = pd.read_csv('x_learn.txt', delimiter = "\t", header=None, names=['text'])
df_test = pd.read_csv('x_test.txt', delimiter = "\t", header=None, names=['text'])
```

```python
df_learn
```

```python
df_test
```

```python
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(df_learn.text)
X_test = vectorizer.transform(df_test.text)
```

```python
df_vectors = pd.DataFrame(cosine_similarity(X_test, X_train), columns = df_learn.index)
df_vectors.to_csv('vectors.csv')
```

```python
df_vectors['idx_test'] = df_test.index
```

```python
df_vectors
```

```python
df_similarity = pd.melt(df_vectors, id_vars='idx_test', var_name='idx_train', value_name='score')
```

```python
df_similarity.sort_values(by=['score'],ascending=False,inplace=True)
df_similarity.reset_index(drop=True,inplace=True)
```

```python
df_similarity[:30]
```

```python
df_sim_text=df_similarity.merge(df_test,how='left', left_on='idx_test', right_on=df_test.index)
df_sim_text.rename(columns={'text':"text_test"},inplace=True)
```

```python
df_sim_text=df_sim_text.merge(df_learn,how='left', left_on='idx_train', right_on=df_learn.index)
df_sim_text.rename(columns={'text':"text_train"},inplace=True)
```

```python
df_sim_text.columns.tolist()
cols = ['idx_test', 'idx_train', 'text_test', 'text_train','score']
```

```python
df_sim_text=df_sim_text[cols]
df_sim_text
```

```python
df_sim_text.to_csv('similarity_with_text.csv',encoding='utf-8-sig')
```
