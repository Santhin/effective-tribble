```python
import pandas as pd
```

```python
a_df = pd.read_csv('a.csv', index_col=0)
a_df.index = pd.to_datetime(a_df.index)
b_df = pd.read_csv('b.csv', index_col=0)
b_df.index = pd.to_datetime(b_df.index)
```

```python
a_df=a_df.resample('M').mean()
b_df=b_df.resample('M').mean()
```

```python
c_df = pd.concat([a_df, b_df], axis=1)
c_df
```

```python
c_df.to_csv("c.csv")
```
